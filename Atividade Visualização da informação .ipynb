## Importando as bibliotecas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##Dataset armazenado na variável dados

```python
dados = pd.read_csv("/netflix_titles.csv") 
```

##Gráfico 1 - Tipo Barras

```python
dados["type"] = dados["type"].replace({"Movie": "Filme", "TV Show": "Série"})
tipos = dados["type"].unique()


quantidade = dados["type"].value_counts()

x_pos = np.arange(len(tipos))
plt.bar(x_pos, quantidade)


plt.xticks(x_pos, tipos)

plt.ylabel("Quantidade")


plt.title("Gráfico de Barras: Quantidade de Filmes e Séries")


plt.show()
```

##Gráfico 2 - Tipo Linha

```python
contagem_por_ano = dados["release_year"].value_counts().sort_index()

plt.plot(contagem_por_ano.index, contagem_por_ano.values)

plt.xlabel("Ano de Lançamento")
plt.ylabel("Quantidade")

plt.title("Gráfico de linhas: Quantidade de lançamentos por ano")

plt.show()
```
